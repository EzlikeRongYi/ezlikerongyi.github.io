## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here
color = {'boxes': 'black', 
         'whiskers': 'green', 
         'medians': 'red', 
         'caps': 'blue'}

df['ConvertedComp'].plot(kind="box", figsize=(8,6), color=color)
plt.grid()
plt.title("Distribution of Compensation")
plt.ylabel("Density")
plt.show()
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
iqr = df['ConvertedComp'].quantile(0.75) - df['ConvertedComp'].quantile(0.25)
print(iqr)
```

Find out the upper and lower bounds.


```python
# your code goes here
upper_limit = df['ConvertedComp'].quantile(0.75) + 1.5 * iqr
lower_limit = df['ConvertedComp'].quantile(0.25) - 1.5 * iqr

print(f"Upper limit ($): {upper_limit}")
print(f"Lower limit ($): {lower_limit}")
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here
outliers = [x for x in df['ConvertedComp'] if x > upper_limit or x < lower_limit]
print(f"Number of outliers: {len(outliers)}")
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
# your code goes here
filtered_data = df.loc[df['ConvertedComp'].isin(outliers) == False]
```

```python
filtered_data['ConvertedComp'].plot(kind="hist", figsize=(10,6), bins=50)
plt.xlim(0, 250_000)
plt.grid()
plt.title("Distribution of Compensation (without outliers))")
plt.xlabel("Compensation (USD)")
plt.ylabel("Density")
plt.show()
```

## Correlation


### Finding correlation


Find the correlation between `Age` and all other numerical columns.


```python
# your code goes here
sns.heatmap(filtered_data.corr(), annot=True)
```
